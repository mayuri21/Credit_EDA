# checking imbalance data

```python
application_data.TARGET.value_counts()
```

```python
plt.pie(application_data.TARGET.value_counts(normalize=False),textprops={'fontsize':16},autopct='%1.f%%',labels=application_data.TARGET.value_counts().index)
plt.title('Pie chart')
```

```python
application_data['TARGET'] = application_data['TARGET'].replace([1],'Defaulter')
application_data['TARGET'] = application_data['TARGET'].replace([0],'Non - Defaulter')
```

```python
application_data['TARGET'].value_counts()
```

```python
application_data['CODE_GENDER'] = application_data['CODE_GENDER'].replace(['M'],'Male')
application_data['CODE_GENDER'] = application_data['CODE_GENDER'].replace(['F'],'Female')
```

```python
application_data['CODE_GENDER'].value_counts()
```

```python
application_data['AMT_INCOME_BIN']=pd.cut(application_data['AMT_INCOME_TOTAL'],bins=[0,25000,50000,75000,100000,125000,150000,175000,200000,225000,250000,275000,
                                                           300000,325000,350000,375000,400000,425000,450000,475000,500000,10000000000],
                                                    labels=['0-25000', '25000-50000','50000-75000','75000-100000','100000-125000', '125000-150000', '150000-175000','175000-200000',
                                                   '200000-225000','225000-250000','250000-275000','275000-300000','300000-325000','325000-350000','350000-375000',
                                                   '375000-400000','400000-425000','425000-450000','450000-475000','475000-500000','500000 and above'])
```

### Unviariate analysis of data

```python
plt.figure(figsize=(5,5))
```

```python
plt.title("Distribution of children count")
```

```python
sns.boxplot(data=application_data, y='CNT_CHILDREN',x='TARGET')
```

```python
plt.figure(figsize=(5,5))
plt.title("Distribution in Credit Amounts",fontsize=20)
sns.boxplot(data=application_data, x='TARGET',y='AMT_ANNUITY', palette='viridis', orient='v')
plt.xlabel("Target Variables", fontsize=15)
plt.ylabel("Credit Amounts", fontsize=15)
plt.show()
```

```python
# Conducting outlier analysis on family members of applicants
plt.figure(figsize=(5,5))
plt.title("Distribution in family members",fontsize=20)
sns.boxplot(data=application_data, x='TARGET',y='CNT_FAM_MEMBERS', palette='viridis', orient='v')
plt.xlabel("Target Variables", fontsize=15)
plt.ylabel("Family Members", fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(20,6))
sns.countplot(y=application_data.AMT_INCOME_BIN, hue=application_data.TARGET, palette='viridis')
plt.title("Income Distribution of Applicants among Income Groups", fontsize=20)
plt.xlabel("Income groups", fontsize=20)
plt.ylabel("Counts", fontsize=20)
plt.xticks(rotation=45,fontsize=12)
plt.legend(prop={'size': 20})
plt.show()
```

```python
plt.figure(figsize=(20,6))
sns.countplot(x=application_data.AMT_INCOME_BIN, hue=application_data.CODE_GENDER,palette='viridis')
plt.title("Income Distribution of Applicants among Genders", fontsize=20)
plt.xlabel("Income groups", fontsize=20)
plt.ylabel("Counts", fontsize=20)
plt.xticks(rotation=45,fontsize=12)
plt.legend(prop={'size': 20})
plt.show()
```

```python
# Plotting a countplot to check distribution of incomes among types of incomes
plt.figure(figsize=(20,6))
sns.countplot(x=application_data.NAME_INCOME_TYPE, hue=application_data.TARGET,palette='viridis')
plt.title("Income Distribution of Applicants among Income Types", fontsize=20)
plt.xlabel("Income Types", fontsize=20)
plt.ylabel("Counts", fontsize=20)
plt.xticks(fontsize=12)
plt.legend(prop={'size': 20},loc='upper right')
plt.show()
```
